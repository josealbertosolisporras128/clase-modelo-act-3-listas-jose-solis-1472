```python
# =========================
# Clase Modelo: MenuLista
# =========================

class MenuLista:

    def __init__(self):
        self.lista = []

    # =========================
    # Agregar
    # =========================
    def agregarProducto(self):
        nombre = input("Ingrese nombre del producto: ")
        precio = float(input("Ingrese precio: "))

        producto = {
            "nombre": nombre,
            "precio": precio
        }

        self.lista.append(producto)
        print("Producto agregado correctamente.\n")

    # =========================
    # Leer
    # =========================
    def leerProducto(self):
        if len(self.lista) == 0:
            print("No hay productos registrados.\n")
            return

        posicion = int(input("Ingrese posición del producto: "))

        if 0 <= posicion < len(self.lista):
            print(self.lista[posicion])
        else:
            print("Posición inválida.\n")

    # =========================
    # Listar
    # =========================
    def listarProductos(self):
        if len(self.lista) == 0:
            print("Lista vacía.\n")
            return

        print("\n===== LISTA DE PRODUCTOS =====")
        for i, producto in enumerate(self.lista):
            print(f"{i} -> Nombre: {producto['nombre']} | Precio: {producto['precio']}")
        print()

    # =========================
    # Buscar
    # =========================
    def buscarProducto(self):
        nombre = input("Ingrese nombre a buscar: ")

        encontrado = False

        for producto in self.lista:
            if producto["nombre"].lower() == nombre.lower():
                print("Producto encontrado:")
                print(producto)
                encontrado = True

        if not encontrado:
            print("Producto no encontrado.\n")

    # =========================
    # Modificar
    # =========================
    def modificarProducto(self):
        nombre = input("Ingrese nombre del producto a modificar: ")

        for producto in self.lista:
            if producto["nombre"].lower() == nombre.lower():

                nuevoNombre = input("Nuevo nombre: ")
                nuevoPrecio = float(input("Nuevo precio: "))

                producto["nombre"] = nuevoNombre
                producto["precio"] = nuevoPrecio

                print("Producto modificado correctamente.\n")
                return

        print("Producto no encontrado.\n")

    # =========================
    # Borrar
    # =========================
    def borrarProducto(self):
        nombre = input("Ingrese nombre del producto a borrar: ")

        for producto in self.lista:
            if producto["nombre"].lower() == nombre.lower():
                self.lista.remove(producto)
                print("Producto eliminado correctamente.\n")
                return

        print("Producto no encontrado.\n")


# =========================
# Menú Principal
# =========================

menu = MenuLista()

while True:

    print("===== MENÚ =====")
    print("1. Agregar producto")
    print("2. Leer producto")
    print("3. Listar productos")
    print("4. Buscar producto")
    print("5. Modificar producto")
    print("6. Borrar producto")
    print("7. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        menu.agregarProducto()

    elif opcion == "2":
        menu.leerProducto()

    elif opcion == "3":
        menu.listarProductos()

    elif opcion == "4":
        menu.buscarProducto()

    elif opcion == "5":
        menu.modificarProducto()

    elif opcion == "6":
        menu.borrarProducto()

    elif opcion == "7":
        print("Saliendo del programa...")
        break

    else:
        print("Opción inválida.\n")
```
